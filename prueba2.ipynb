```python
# Carga del dataframe con el archivo user_reviews.json.gz
data = []
with gzip.open('dataset/user_reviews.json.gz', 'rt', encoding='utf-8') as file:
    for line in file:
        try:
            # Suponiendo que las líneas contienen datos JSON válidos
            json_data = eval(line.strip())  # Suponiendo una estructura JSON simple, use ast.literal_eval() para mayor seguridad
            data.append(json_data)
        except ValueError as e:
            print(f"Error en la línea: {line}")
            continue

df_user_reviews = pd.DataFrame(data)
```

```python
# Observamos la estructura del archivo
df_user_reviews.info()
df_user_reviews.head()
```

```python
# Mostrar los nombres de las columnas disponibles del DataFrame user_reviews
print(df_user_reviews.columns)

# Mostrar sus tipos de datos
for column in df_user_reviews.columns:
    print(f"Etiqueta: {column}, Tipo de dato: {df_user_reviews[column].dtype}")
```

```python
# Verificar duplicados
tiene_duplicados = df_user_reviews['user_id'].duplicated().any()
duplicados = df_user_reviews[df_user_reviews['user_id'].duplicated()]['user_id']
print(duplicados)
```

```python
# Eliminar duplicados y mantener la primera aparición

df_user_reviews = df_user_reviews.drop_duplicates(subset=['user_id'], keep='first')

df_user_reviews
```

```python
# Verificar duplicados
tiene_duplicados = df_user_reviews['user_id'].duplicated().any()
duplicados = df_user_reviews[df_user_reviews['user_id'].duplicated()]['user_id']
print(duplicados)
```

```python
# Transformación de la columna reviuws que se requiere para las consultas solicitadas

#con la funcion explode() se desglosa las listas de diccionarios en filas individuales
copia = df_user_reviews.explode('reviews') 

# concatena para que cada item del diccionario quede en columnas individuales.
copia = pd.concat([copia.drop(['reviews'],axis= 1), copia['reviews'].apply(pd.Series)], axis = 1) 
copia.rename(columns= {0: 'sentiment_analysis'}, inplace= True)

# se borra las columnas que no son necesarias
copia.drop(['funny', 'posted', 'last_edited', 'helpful'], axis=1, inplace=True) 
copia['sentiment_analysis'] = 0
copia.head()
```

```python
#Revision de valores nulos para 'item_id'

copia['item_id'].isna().value_counts()
```

```python
# Eliminar los nulos

copia = copia.dropna(subset= ['item_id'])
copia.shape
copia.info()
```

```python
"""  En la colunmna o etiqueta sentiment_analysis se encuentra con valore 0, se decide actualizar estos valores aplicando por medio de la 
 librearia textlob para análisis de sentimientos con NLP, que como se indica en la rubrica, se debe aplicar la siguiente escala:
 debe tomar el valor '0' si es malo, '1' si es neutral y '2' si es positivo. Lo anterior para facilitar el trabajo de los modelos
 de machine learning y el análisis de datos """
    
from textblob import TextBlob

# Aplicar el análisis de sentimiento directamente
copia['sentiment_analysis'] = copia['review'].apply(
    
    
    lambda review: 1 if not review else (lambda analisis: 0 if analisis.sentiment.polarity < 0 else (1 if analisis.sentiment.polarity == 0 else 2))(TextBlob(review))
)

copia
```

```python
# Eliminar columnas innecesarias
copia = copia.drop(columns=['user_url', 'review'])
copia
```

```python
# Finalizando, se exportar el dataset copia a formato parquet
copia.to_parquet('dataset/df_user_reviews.parquet', engine='pyarrow')
```
